```python
def rb_tree_insert(T, z):
    y = T.nil
    x = T.root
    while x != T.nil:
        y = x
        if z.key < x.key:
            x = x.left
        else:
            x = x.right
    z.p = y
    if y == T.nil:
        T.root = z
    elif z.key < y.key:
        y.left = z
    else:
        y.right = z
    rb_insert_fixup(T, z)
```

```python
def left_rotate(T, x):
    y = x.right
    x.right = y.left
    if y.left != T.nil:
        y.left.p = x
    y.p = x.p
    if x.p == T.nil:
        T.root = y
    elif x == x.p.left:
        x.p.left = y
    else:
        x.p.right = y
    y.left = x
    x.p = y

def right_rotate(T, y):
    x = y.left
    y.left = x.right
    if x.right != T.nil:
        x.right.p = y
    x.p = y.p
    if y.p == T.nil:
        T.root = x
    elif y == y.p.left:
        y.p.left = x
    else:
        y.p.right = x
    x.right = y
    y.p = x
```

```python
def rb_transplant(T, u, v):
    if u.p == T.nil:
        # u is root of T
        T.root = v
    elif u == u.p.left:
        u.p.left = v
    else:
        u.p.right = v
    v.p = u.p

def tree_min(x):
    if not x.left:
        return x
    else:
        return tree_min(x.left)
    
def rb_tree_delete(T, z):
    y = z
    y_original_color = y.color
    if z.left == T.nil:
        x = z.right
        rb_transplant(T, z, z.right)
    elif z.right == T.nil:
        x = z.left
        rb_transplant(T, z, z.left)
    else:
        y = tree_min(z.right)
        y_original_color = y.color
        x = y.right
        if y.p == z:
            x.p = y
        else:
            rb_transplant(T, y, y.right)
            y.right = z.right
            y.right.p = y
        rb_transplant(T, z, y)
        y.left = z.left
        y.left.p = y
        y.color = z.color
    if y_original_color == black:
        rb_delete_fixup(T, x)

def rb_delete_fixup(T, x):
    while x != T.root and x.color == black:
        if x == x.p.left:
            w = x.p.right
            if w.color == red:
                x.p.color = red
                w.color = black
                left_rotate(T, x.p)
                w = x.p.right
            if w.left.color == black and w.right.color == black:
                w.color = red
                x = x.p
            else:
                if w.right.color == black:
                    w.left.color = black
                    w.color = red
                    right_rotate(T, w)
                    w = x.p.right
                w.color = x.p.color
                w.right.color = black
                x.p.color = black
                left_rotate(T, x.p)
                x = T.root
        else:
            w = x.p.left
            if w.color == red:
                x.p.color = red
                w.color = black
                right_rotate(T, x.p)
                w = x.p.left
            if w.left.color == black and w.right.color == black:
                w.color = red
                x = x.p
            else:
                if w.left.color == black:
                    w.right.color = black
                    w.color = red
                    left_rotate(T, w)
                    w = x.p.left
                w.color = x.p.color
                x.p.color = black
                w.left.color = black
                right_rotate(T, x.p)
                x = T.root
    x.color = black
```
