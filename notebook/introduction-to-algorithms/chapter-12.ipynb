```python
class Tree:
    def __init__(self):
        self.root = None;
    
    def as_json_string(self):
        if self.root != None:
            return self.root.as_json_string()
        else:
            return ''

class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.p = None
        self.key = key
    
    def __str__(self):
        ret = 'key = ' + str(self.key)
        if self.p:
            ret = ret + ', parent = ' + str(self.p.key)
        if self.left:
            ret = ret + ', left = ' + str(self.left.key)
        if self.right:
            ret = ret + ', right = ' + str(self.right.key)
        return ret
        
    def as_json_string(self):
        ret = '{"key":' + str(self.key)
        if self.left:
            ret = ret + ',"left":' + self.left.as_json_string()
        if self.right:
            ret = ret + ',"right":' + self.right.as_json_string()
        ret = ret + '}'
        return ret
        
    def insert(self, key):
        if key < self.key:
            if self.left == None:
                left = Node(key)
                left.p = self
                self.left = left
                return self.left
            else:
                self.left.insert(key)
        if key > self.key:
            if self.right == None:
                right = Node(key)
                right.p = self
                self.right = right
                return self.right
            else:
                self.right.insert(key)
        return self
```

```python
def tree_insert(T, z):
    y = None
    x = T.root
    while x != None:
        y = x
        if z.key < x.key:
            x = x.left
        else:
            x = x.right
    z.p = y
    if y == None:
        T.root = z
    else:
        if z.key < y.key:
            y.left = z
        else:
            y.right = z
```

```python
T1 = Tree()
n15 = Node(15)
n6 = Node(6)
n3 = Node(3)
n2 = Node(2)
n4 = Node(4)
n7 = Node(7)
n13 = Node(13)
n9 = Node(9)
n18 = Node(18)
n17 = Node(17)
n20 = Node(20)

tree_insert(T1, n15)
tree_insert(T1, n6)
tree_insert(T1, n3)
tree_insert(T1, n2)
tree_insert(T1, n4)
tree_insert(T1, n7)
tree_insert(T1, n13)
tree_insert(T1, n9)
tree_insert(T1, n18)
tree_insert(T1, n17)
tree_insert(T1, n20)

print(T1.as_json_string())
```

```python
def tree_search(T, k):
    y = None
    x = T.root
    while x != None:
        if k == x.key:
            y = x
            x = None
        elif k < x.key:
            x = x.left
        else:
            x = x.right
    return y

n = tree_search(T1, 6)

print(n)
```

```python
def tree_min(x):
    if x.left == None:
        return x
    else:
        return tree_min(x.left)
print(tree_min(T1.root))
```

```python
def tree_max(x):
    if x.right == None:
        return x
    else:
        return tree_max(x.right)
print(tree_max(T1.root))
```

```python
def tree_successor(x):
    if x.right:
        return tree_min(x.right)
    y = x.p
    while y and x == y.right:
        x = y
        y = y.p
    return y

n15 = tree_search(T1, 15)
print(tree_successor(n15))

n13 = tree_search(T1, 13)
print(tree_successor(n13))
```

```python
def transplant(T, u, v):
    if u.p == None:
        # u is root of T
        T.root = v
    elif u == u.p.left:
        u.p.left = v
    else:
        u.p.right = v
        
def tree_delete(T, z):
    if z.left == None:
        transplant(T, z, z.right)
    elif z.right == None:
        transplant(T, z, z.left)
    else:
        y = tree_min(z.right)
        if y.p != z:
            transplant(T, y, y.right)
            y.right = z.right
            y.right.p = y
        transplant(T, z, y)
        y.left = z.left
        y.left.p = y
```
